Import Libraries

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
```

Load Data

```python
trainData = pd.read_csv('train_V2.csv')
testData = pd.read_csv('test_V2.csv')
print(trainData.shape, testData.shape)
```

```python
# Merge all data together to ahve better data set
pubgData = pd.concat([trainData, testData], axis =0)
print(pubgData.shape)
pubgData.head()
```

```python
pubgData.info()
#none of the input has Null data
```

```python
pubgData.describe().T
```

```python
# Drop any null values if any
# dropna
pubgData.dropna(inplace=True)

nullCount = pubgData.isnull().sum().sort_values()
print('Null count : ', nullCount[nullCount > 0])
```

Plot kill counts to see any pattern

Remove unwanted inputs like all IDs

```python
modelData = pubgData.drop(['Id', 'groupId', 'matchId', 'matchType'], axis = 1)
modelData.head()
```

```python
X = modelData.drop(['winPoints'], axis=1).values
y = modelData['winPoints'].values
```

Split into train and test data 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
print((X_train.shape))
print((X_test.shape))
print((y_train.shape))
print((y_test.shape))
```

Apply normalization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

Build ANN using Sequential and Layers class

```python
from model_functions import model_train
ann = model_train(X_train, y_train, 12, 'relu', 50, 100)
```

Predicting test results

```python
y_pred = ann.predict(X_test)
y_pred
```
